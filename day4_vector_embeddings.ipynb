```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

sentences = [
    "The voter registration process for the upcoming Indian election has begun.",
    "Citizens can now enroll to cast their ballots in the national polls.",
    "Deep learning models can accurately classify different breeds of livestock.",
    "Farmers use computer vision to identify cow and poultry variations.",
    "The weather forecast predicts heavy rain across the region."
]

embeddings = model.encode(sentences)

score_1_2 = util.cos_sim(embeddings[0], embeddings[1])[0][0].item()
score_3_4 = util.cos_sim(embeddings[2], embeddings[3])[0][0].item()
score_1_5 = util.cos_sim(embeddings[0], embeddings[4])[0][0].item()

print(f"Similarity (Elections 1 & 2): {score_1_2:.4f}")
print(f"Similarity (Livestock 3 & 4): {score_3_4:.4f}")
print(f"Similarity (Elections & Weather 1 & 5): {score_1_5:.4f}")
```

```python
from sentence_transformers import util

# Similarity between sentence 1 and 2 (Election-related)
score_1_2 = util.cos_sim(embeddings[0], embeddings[1])

# Similarity between sentence 3 and 4 (AI/Farming-related)
score_3_4 = util.cos_sim(embeddings[2], embeddings[3])

# Similarity between sentence 1 and 5 (Different meanings)
score_1_5 = util.cos_sim(embeddings[0], embeddings[4])

print(f"Similarity (Elections 1 & 2): {score_1_2.item():.4f}")
print(f"Similarity (AI/Farming 3 & 4): {score_3_4.item():.4f}")
print(f"Similarity (Elections & Weather 1 & 5): {score_1_5.item():.4f}")
```

## Reflection

I was surprised that the model could detect similar meanings even when the wording was completely different.
The election-related sentences had a high similarity score because both discussed voting and registration.
The AI and livestock sentences were also semantically related despite using different vocabulary.
The weather sentence showed a much lower similarity score, which matched my intuition.
This experiment helped me understand how embeddings capture meaning in vector space.
