```python
import re
from collections import Counter
import nltk

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
text = """
Artificial Intelligence is transforming the world!
AI systems can understand language, analyze images, and automate decisions.
However, raw text often contains punctuation, extra spaces, and inconsistent casing.
"""
```

```python
cleaned = text.lower()
cleaned = re.sub(r'[^\w\s]', '', cleaned)
cleaned = " ".join(cleaned.split())

print(cleaned)
```

```python
tokens = word_tokenize(cleaned)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print(filtered_tokens)
```

```python
freq = Counter(filtered_tokens)

print(freq)
```

# Engineering Observations

Raw text is often noisy and inconsistent, so preprocessing is essential before AI systems can use it effectively.

Converting text to lowercase helps maintain consistency.

Removing punctuation and extra spaces makes the data cleaner and easier to process.

Tokenization breaks text into smaller machine-readable units.

Stopword removal helps focus on meaningful words, improving downstream NLP performance.
