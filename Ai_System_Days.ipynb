```python
text = input("Enter a sentence: ")

words = text.lower().split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)
```

```python
import re

def clean_text(raw_text):
    text = raw_text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

messy_input = "   HELLO!!!   This is Day 2 of AI...   Let's clean this text up.   "
print("Cleaned Text:", clean_text(messy_input))
```
