```python
import re
import nltk
from nltk.corpus import stopwords
from textblob import TextBlob

nltk.download('stopwords')
```

```python
text = """
I absolutely love learning Artificial Intelligence.
The projects are exciting, but debugging can sometimes be frustrating.
Still, building AI systems feels rewarding.
"""
```

```python
cleaned = text.lower()
cleaned = re.sub(r'[^\w\s]', '', cleaned)
cleaned = " ".join(cleaned.split())

print("Cleaned Text:")
print(cleaned)
```

```python
tokens = cleaned.split()

stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]

print("Filtered Tokens:")
print(filtered_tokens)
```

```python
blob = TextBlob(cleaned)

print("Sentiment Polarity:", blob.sentiment.polarity)
print("Sentiment Subjectivity:", blob.sentiment.subjectivity)
```

```python
def nlp_pipeline(text):
    cleaned = text.lower()
    cleaned = re.sub(r'[^\w\s]', '', cleaned)
    cleaned = " ".join(cleaned.split())

    tokens = cleaned.split()
    filtered = [word for word in tokens if word not in stop_words]

    sentiment = TextBlob(cleaned).sentiment

    return {
        "cleaned_text": cleaned,
        "tokens": filtered,
        "polarity": sentiment.polarity,
        "subjectivity": sentiment.subjectivity
    }

sample = "AI development is challenging but incredibly exciting."

result = nlp_pipeline(sample)

print(result)
```

# Engineering Observations

An NLP pipeline transforms raw human text into structured machine-processable data.

Cleaning improves consistency and removes noise.

Tokenization converts text into smaller units for analysis.

Stopword removal helps focus on meaningful information.

Sentiment analysis demonstrates how AI can extract insights from text automatically.
