```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

dataset = [
    "Artificial intelligence helps automate repetitive tasks.",
    "Machine learning is a branch of AI.",
    "Deep learning uses neural networks for complex tasks.",
    "Natural language processing helps machines understand text.",
    "Computer vision allows AI to analyze images.",
    "Chatbots use AI to communicate with users.",
    "Recommendation systems suggest products or content.",
    "Semantic search understands meaning instead of exact words.",
    "RAG combines retrieval with language generation.",
    "Vector databases store embeddings for fast retrieval.",
    "Python is commonly used for AI development.",
    "FastAPI helps deploy AI applications as APIs.",
    "Gemini API enables developers to build AI-powered apps.",
    "Transformers revolutionized NLP tasks.",
    "Embeddings convert text into numerical vectors.",
    "Cosine similarity measures semantic closeness.",
    "Speech recognition converts voice into text.",
    "AI assistants answer questions intelligently.",
    "Data preprocessing improves model performance.",
    "Neural networks learn patterns from data."
]
```

```python
embeddings = model.encode(dataset)

print("Dataset embedded successfully!")
```

```python
def semantic_search(query, top_k=3):
    query_embedding = model.encode(query)

    scores = util.cos_sim(query_embedding, embeddings)[0]

    results = []

    for i, score in enumerate(scores):
        results.append((dataset[i], score.item()))

    results = sorted(results, key=lambda x: x[1], reverse=True)

    return results[:top_k]
```

```python
query = "How do AI systems understand human language?"

results = semantic_search(query)

print("Query:", query)
print("\nTop 3 Results:\n")

for sentence, score in results:
    print(f"{sentence} | Score: {score:.4f}")
```

```python
def keyword_search(query):
    query_words = query.lower().split()

    matches = []

    for sentence in dataset:
        count = sum(word in sentence.lower() for word in query_words)
        matches.append((sentence, count))

    matches = sorted(matches, key=lambda x: x[1], reverse=True)

    return matches[:3]
```

```python
print("SEMANTIC SEARCH RESULTS:\n")
semantic_results = semantic_search(query)

for sentence, score in semantic_results:
    print(sentence)

print("\nKEYWORD SEARCH RESULTS:\n")
keyword_results = keyword_search(query)

for sentence, count in keyword_results:
    print(sentence)
```

## Comparison

Semantic search performed better because it understood the meaning of the query rather than relying only on exact keyword matches.

Keyword search only looked for matching words, which sometimes missed relevant results.

Semantic search returned contextually accurate results, making it more useful for modern AI applications like RAG and recommendation systems.
